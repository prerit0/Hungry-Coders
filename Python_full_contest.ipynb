**Q1**. MedCare wants to analyze the effectiveness of their social media campaigns. They have a dictionary with social media platforms and their follower counts. The task is to reverse the platform name only if the follower count is odd. For example, for 15 followers on Twitter, the name should be reversed to "rettiwT". If the follower count is even, the name remains unchanged.

Write a Python function that processes the dictionary and returns a list of platform names, reversing only those with an odd follower count.

```python
def reverse_odd_platforms(data):
    result = []
    for platform, followers in data.items():
        if followers % 2 != 0:
            result.append(platform[::-1]) # reverse name if odd
        else:
            result.append(platform)       # keep as is
    return result


# Example
followers_dict = {
    'Instagram': 234,
    'Twitter': 223,
    'LinkedIn': 456
}
print(reverse_odd_platforms(followers_dict))
```

**Q2**. Write a Python function that accepts two NumPy arrays (representing matrices) as input and performs matrix multiplication on them. The function should return the resulting matrix. Use the np.dot() function to compute the dot product of the two matrices. Ensure the dimensions of the input matrices are compatible for multiplication. The function should handle any valid matrix sizes, where the number of columns in the first matrix matches the number of rows in the second matrix.

```python
import numpy as np

def matrix_multiply(a, b):
    # Ensure matrices are compatible
    if a.shape[1] != b.shape[0]:
        raise ValueError("Incompatible dimensions for matrix multiplication.")

    result = np.dot(a, b)
    return result


# Example
A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

print(matrix_multiply(A, B))
```

**Q3**. "SocialConnect" wants to analyze user engagement on their platform. Given a DataFrame with columns like 'user_id', 'post_id', 'engagement_type' (like, comment, share), and 'timestamp'.

```python
def find_top_engagement(df):
    return df['engagement_type'].value_counts().head()


# Example DataFrame
import pandas as pd

df = pd.DataFrame({
    'user_id': [1, 2, 3, 4, 5],
    'post_id': ['a', 'b', 'c', 'd', 'e'],
    'engagement_type': ['like', 'comment', 'share', 'like', 'share'],
    'timestamp': ['10:22:29', '10:50:29', '08:01:15', '06:45:39', '12:55:55']
})

print(find_top_engagement(df))
```

  **Q4**. A health and wellness company, FitBite, collects data on the number of daily steps taken by their users. They have two lists of values, steps_day1 and steps_day2, which represent the number of steps taken by a user on day 1 and day 2, respectively. The company wants to visualize this data to identify any patterns or trends. Write a Python program using Matplotlib that generates a line plot of these two lists of values, then customizes the plot by adding a title, labels for the x and y axes, and a grid.

The steps_day1 list contains: [1000, 1200, 1100, 1300, 1400] The steps_day2 list contains: [900, 1000, 1200, 1100, 1300]

Your task is to write a Python program that generates the plot chart.

```python
import matplotlib.pyplot as plt

steps_day1 = [1000, 1200, 1100, 1300, 1400]
steps_day2 = [900, 1000, 1200, 1100, 1300]

days = [1, 2, 3, 4, 5]  # x-axis representing days

plt.plot(days, steps_day1, label='Day 1 Steps', marker='o')
plt.plot(days, steps_day2, label='Day 2 Steps', marker='s')

plt.title('Daily Steps Comparison')
plt.xlabel('Day')
plt.ylabel('Number of Steps')
plt.xticks(days)
plt.grid(True)
plt.legend()
plt.show()
```

**Q5**. A healthcare company, MedCorp, has a dataset containing information about patients, including their ages, blood pressures, and disease statuses (e.g., diabetes, hypertension). The company wants to visualize the relationships between these variables using a pair plot with categorical color coding. The dataset is stored in a CSV file called "patients_data.csv" and has the following structure:

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# df = pd.read_csv('patients_data.csv')
df = pd.DataFrame({
    'Age': [25, 30, 28, 35, 40, 45, 50, 32, 38, 42],
    'Blood_Pressure': [120, 130, 125, 140, 135, 150, 155, 128, 138, 145],
    'Disease_Status': [
        'Diabetes', 'Hypertension', 'Diabetes', 'Hypertension',
        'Diabetes', 'Hypertension', 'Hypertension', 'Diabetes',
        'Diabetes', 'Hypertension'
    ]
})

sns.pairplot(df, hue='Disease_Status')
plt.suptitle('Pair Plot of Patient Data by Disease Status', y=1.02)
plt.show()
```
