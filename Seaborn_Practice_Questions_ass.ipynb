## Question 1: Comparing Two Lists Using a Line Plot

Problem Statement:
You are given two lists representing monthly sales of two products over 6 months. 
You need to visualize how their sales trends compare over time using a line plot.

Approach:
1. Create two lists: months and sales.
2. Use `sns.lineplot()` to plot both lines on the same graph.
3. Add appropriate labels and a title to make the visualization clear.

```python
# Write your code here
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example lists
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_A = [45, 55, 60, 70, 65, 75]
sales_B = [40, 50, 58, 68, 70, 80]

# Your line plot code here
# df = pd.concat('months': 'months')
sns.lineplot(x = months, y = sales_A,  label = 'sales A', color = 'red', errorbar = None) # instead of ci we use errorbar because ci is deprecated
sns.lineplot(x = months, y = sales_B, label = 'sales B' , color = 'black')
plt.title('Monthly sales report')
plt.xlabel('months')
plt.ylabel('sales')
plt.show()
```

## Question 5: Average Fare by Day of the Week

Problem Statement:
Do taxi fares vary across different days of the week?

Approach:
1. Use the 'pickup_day' or derive day information from the datetime column if available.
2. Plot the average fare per day using `sns.barplot()`.

```python
# Write your code here
taxis['pickup'] = pd.to_datetime(taxis['pickup'])  #Convert pickup datetime to datetime type
taxis['pickup_day'] = taxis['pickup'].dt.day_name() #Derive the day of the week

avg_fare = taxis.groupby('pickup_day')['fare'].mean().reset_index()

days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.figure(figsize=(8,5))
sns.barplot(x='pickup_day', y='fare',data = avg_fare, order = days_order, hue = 'pickup_day')
plt.title('Average Taxi Fare by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Fare')
plt.show()
```

## Question 8: Trip Duration vs. Fare Amount

## Question 10: Average Tip by Distance Category

Problem Statement:
We want to know if longer trips generally result in higher tips.

Approach:
1. Create distance categories (e.g., short, medium, long) using the 'distance' column.
2. Use `sns.barplot()` to plot average tips for each distance category.

```python
# Write your code here
bins = [0, 2, 5, taxis['distance'].max()]  # Short: 02, Medium: 25, Long: 5+
labels = ['Short', 'Medium', 'Long']
taxis['distance_category'] = pd.cut(taxis['distance'], bins=bins, labels=labels)
sns.barplot(data=taxis, x='distance_category', y='tip', color = 'grey', edgecolor = 'black')
plt.title('Average Tip Amount by Trip Distance Category')
plt.xlabel('Trip Distance Category')
plt.ylabel('Average Tip')
plt.show()
```
